## Exercise 5

As for the distributions implemented in R, four function can be defined. 

Starting from the pdf function,  the normalization factor is given by the formula: $$ c = \frac{1}{\int_1^2 \left(x-1\right)\left(2-x\right) dx} = 6$$

```r
dtolc = function(x,c=6) {
    #Returns the pdf computed in x
    ifelse((x>1) & (x<2), c*(x-1)*(2-x), 0) 
}
```

Using the `integrate()` function, is possible to check if the correct normalization factor is $c=6$.

```r
x.integral = integrate(dtolc, 1, 2, c=1)
```

```r
message("The integration result is ", x.integral$value, " with error ", format(x.integral$abs.error, digits = 4))
message("The normalization factor obtained numerically is ", 1/x.integral$value)
```

The other functions can now be defined. In particular the cdf can be computed analitically as

\begin{equation}
  cdf(x) =
    \begin{cases}
      0 & x<1\\
      \int_1^x c\left(x-1\right)\left(2-x\right)dx = c\left(-\frac{x^3}{3}+\frac{3x}{2} -2x -\frac{5}{6}\right) & 1<x<2\\
      1 & x>2
    \end{cases}       
\end{equation}

```r
ptolc = function(x,c=6){
    # Returns the cdf computed in x
    ifelse((x>1) & (x<2), c*(-(x^3)/3+3/2*x^2-2*x+5/6), ifelse(x<=1,0,1))
}
```

Since it is not possible to find analitically the inverse cumulative function, `uroot()` is used in order to find computationally the corresponding quantile for each given probability.

```r
invcdf = function (x) uniroot(function(y){ptolc(y) - x}, lower = 1, upper = 2)$root
```

```r
qtolc = function(x,c=6){
    # Returns the quantile corresponding to the probability given
    ifelse((x>=0) & (x <=1),sapply(x,invcdf),NA)
}
```

```r
rtolc = function(n,c=6){
    # Returns a set of simulated values, drawn from the distribution
    us = runif(n)
    return (ifelse((us>=0) & (us <=1),sapply(us,invcdf),NA))
}
```

```r
plot(seq(0,3,0.01),dtolc(seq(0,3,0.01),), xlab = "Time to end the test [hours]", ylab = "Probability density", main = "Probability distribution", type='l')
```

```r
plot(seq(0,3,0.01),ptolc(seq(0,3,0.01),), xlab = "Time to end the test [hours]", ylab = "Probability", main = "Cumulative probability distribution", type='l')
```

The probability to do the exam in more than 75 minutes is obtained as the difference between 1 and the probability to do the exam in less than 75 minutes.

```r
message("The probability to do the exam in more than 75 minutes is ",format(1-ptolc(5/4), digits = 4))
```

In the same way the probability to do the exam in a time between 75 and 120 minutes can be obtained using the cumulative distribution function.

```r
message("The probability to do the exam in a time between 90 and 120 minutes is ",format(ptolc(2)-ptolc(1.5), digits = 4))
```

## Exercise 6

As for the previous exercise the four function can be defined by using the pdf, the cdf integrated analitically and the inverse cumulative function, that this time can be computed analitically too.

In particular, the analitical expression for the cdf is
\begin{equation}
  cdf(x) =
    \begin{cases}
      0 & x<1\\
      \int_1^x \frac{2}{x^2}= -\frac{2}{x} +2 & 1<x<2\\
      1 & x>2
    \end{cases}       
\end{equation}


```r
dtire = function(x,c=6) {
    #Returns the pdf computed in x
    ifelse((x>1) & (x<2), 2/x^2, 0)
}
```

```r
ptire = function(x,c=6){
    #Returns the cdf computed in x
    ifelse((x>1) & (x<2), -2/x+2, ifelse(x<=1,0,1))
}
```

```r
qtire = function(x,c=6){
    # Returns the quantile corresponding to the probability given
    2/(-x+2)
}
```

```r
rtire = function(n,c=6){
    # Returns a set of simulated values, drawn from the distribution
    us = runif(n)
    return(2/(-us+2))  
}
```

```r
plot(seq(0,3,0.01),dtire(seq(0,3,0.01),), xlab = expression("Tire lifetime/10^4 [km]"), ylab = "Probability density", main = "Probability distribution", type='l')
```

```r
plot(seq(0,3,0.01),ptire(seq(0,3,0.01),), xlab = expression("Tire lifetime/10^4 [km]"), ylab = "Probability", main = "Cumulative probability distribution", type='l')
```

```r
message("The probability that tires will last less than 15000 km is ",format(ptire(1.5), digits=4))
```
